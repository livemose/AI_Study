```python
import pandas as pd
import numpy as np
from pandas_datareader import data

from bs4 import BeautifulSoup as bsf 
import requests 
```

```python
url = 'https://finance.naver.com/sise/sise_group_detail.nhn?type=upjong&no=235'
response = requests.get(url) 
response
soup = bsf(response.text, 'html.parser')
lists = soup.findAll('td', {'style' : 'padding-left:10px;'}) 

type_list = []
type_url_list = []

for item in lists:
    name = item.text
    #url주소
    url = item.find('a').get('href')
    type_list.append(name)
    type_url_list.append(url)
```
