# selenium을 활용

* selenium: 동적으로 화면을 업데이트
* 크롬창을 직접 오픈해서 html 페이지를 받아옴

```python
# (실습3) 이미지 파일을 받아오고 싶어서, 그냥 requests로 하면 안된다...
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr'
seoul_url = '/Restaurants-g294197-Seoul.html'
```

```python
# get html

## 크롬창을 직접
#options = Options()
# options.headless = True
# browser = webdriver.Chrome() 원래 이건데
browser = webdriver.Chrome(options=options)  # 이렇게 하면 크롬이 열리지 않고 백그라운드에서 알아서 해줌
browser.implicitly_wait(3)  # 3초 정도 기다리고 해라
browser.get(base_url+seoul_url)  # url 가져와
time.sleep(5)  # 5초 있다가 해라
# BeautifulSoup을 활용하여 데이터 파싱
html = browser.page_source   # page_source는 그냥 html만 보여주는 것 
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 이미지
photos = soup.findAll('div', {'class' : '_2Hb-Mt7l'})
# print(photos)   # 값이 있는게 보인다
```

```python
for item in photos:
    print(item)
    print('---')
```

```python
img_dir = f"{os.getcwd()}/img"
if not os.path.isdir(img_dir):
   os.mkdir(img_dir)

for photo in photos :
    url = photo.get('style').split('"')[1]
    #print(url)
    img_file_name = url.split('/')[-1]     # / 이걸 기준으로 분류하고 맨 뒤에꺼만 호출
    img_data = requests.get(url).content
    with open(f'img/i{img_file_name}', 'wb') as handler:  # img 파일을 열어서 img_data를 쓰겠다
        handler.write(img_data)
```
