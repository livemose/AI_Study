 ## 불리언 값으로 선택하기

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.randn(7,4)

names
```

```python
data
```

```python
names =='Bob'         # Bob인 애들만 True 나머지 False, 각각의 이름은 data 배열의 각 로우와 대응
```

```python
data[names == 'Bob']      # 불리언 배열은 색인하려는 축의 길이와 같아야함, 그렇지 않으면 나머지는 삭제함
```

```python
data[names == 'Bob', 2:]   # names =='Bob' 인 로우에서 2: 컬럼을 선택했다. 
```

```python
data[names == 'Bob', 3]
```

```python
names !='Bob'
```

```python
data[~(names == 'Bob')]        ~ 를 사용하면 조건절 부인을 의미함, 
```

```python
cond = names =='Bob'

data[~cond]
```

```python
mask = (names == 'Bob') | (names== 'Will')    # |은 or임 , &가 and
mask
```

```python
data[mask]
```

```python
data[data<0] =0          # 이런것도 된다...
data
```

## 배열 데이터 파일 입출력

```python
arr = np.arange(10)
np.save('some_array', arr)       # some_array 가 npy 형태로 저장되어 있음, 안의 내용은 arr
```

```python
np.load('some_array.npy')        # load는 파일을 불러옴 
```

```python
np.savez('array_archive.npz', a=arr, b=arr)   # savez는 압축된 형식으로 저장 할 수 있음
```

```python
arch = np.load('array_archive.npz')

arch['b']
```

```python
np.savez_compressed('arrays_compressed.npz', a=arr, b=arr)
```
