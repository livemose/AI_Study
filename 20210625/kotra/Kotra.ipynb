### 해당 연도 해당 국가의 해당 품목 수입금액 결측치 찾기

```python
products=list(trade_df[trade_df['해당 연도 해당 국가의 해당 품목 수입금액'].isnull()]['품목코드'])
```

```python
trade_df[trade_df['해당 연도 해당 국가의 해당 품목 수입금액'].isnull()]['품목코드']
```

```python
products=set(products)
products
```

```python
trade_df[trade_df['품목코드']==190219]['해당 연도 해당 국가의 해당 품목 수입금액'].median()
```

```python
trade_df[trade_df['해당 연도 해당 국가의 해당 품목 수입금액'].isnull()]['국가명'].value_counts()
```

```python
# 국가마다 해당 품목 수입금액의 중간값을 결측값에 대입

for product in products:
    condition1 = condition1 = (trade_df['해당 연도 해당 국가의 해당 품목 수입금액'].isna())
    
    condition2 = (trade_df['품목코드'] == product)
    
    trade_df.loc[condition1 & condition2, ['해당 연도 해당 국가의 해당 품목 수입금액']] = trade_df[trade_df['품목코드']==product]['해당 연도 해당 국가의 해당 품목 수입금액'].median()
```
